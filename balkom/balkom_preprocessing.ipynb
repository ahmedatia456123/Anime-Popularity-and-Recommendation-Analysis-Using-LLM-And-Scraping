## Extracting Views

```python
df_veiws = df.copy()
```

```python
def extracting_veiws(text):
    return int(text.split(' ')[0])
```

```python
extracting_veiws(df_veiws.iloc[0].veiws)
```

```python
df_veiws.veiws = df_veiws.veiws.apply(extracting_veiws)
```

```python
df_veiws.head(1)
```

## Extracting episode_number

```python
df_episode_number = df_veiws.copy()
```

```python
def extracting_episode_number(text):
    
    try:
        return text.split(':')[1].strip()
    except AttributeError:
        return '1'
```

```python
extracting_episode_number(df_veiws.iloc[1000].episode_number)
```

```python
df_episode_number.episode_number = df_veiws.episode_number.apply(extracting_episode_number)
```

## Grouping

```python
df_gropued = df_episode_number.copy()
```

```python
df_gropued = df_gropued.sort_values('episode_number').groupby('name',group_keys=True)[['rating','myanimelist','veiws','counting_comment','episode_number']].agg(list).reset_index()
```

```python
df_gropued.head()
```

```python
def correcting_raw(raw):
    return raw[0]
```

```python
wanted_column = ['rating','myanimelist']
for c in wanted_column:
    df_gropued[c] = df_gropued[c].apply(correcting_raw)
```

Checking Null Values if there any i will delete them

```python
df_gropued.dropna(subset='myanimelist',inplace=True)
```

Function to get Id and return None if the link not following this structure which means its not MAL link

```python
def get_ids(x):
    
    try:
        return int(x.split('anime/')[1].split('/')[0])
    except:
        return None
```

```python
df_gropued['myanimelist_id'] = df_gropued['myanimelist'].apply(get_ids)
```

Creating mean for views and comments 

```python
from statistics import mean
df_gropued['mean_views'] = df_gropued.veiws.apply(lambda x:mean(x))
df_gropued['comments'] = df_gropued.counting_comment.apply(lambda x:mean(x))
df_gropued['episode_number'] = df_gropued.episode_number.apply(lambda x:max(x))
```

```python
df_gropued.to_csv('balkom_processed.csv',index=False)
```
